## Lendo e propondo análises para o banco de dados do IBGE

Está na hora de colocar em prática todos os conhecimentos aprendidos em aula através de um desafio.

A proposta do desafio 1 é trabalhar a situação da elaboração de questões e a resolução destas questões dentro de um banco de dados conhecido. As questões podem ser consideradas hipóteses e a proposta é elaborar problemas que sejam resolvidos com a análise do conhecido DataSet do IBGE, que estamos trabalhando em nossas aulas. Como um passo a passo para esta atividade, consideramos:

1. Abertura e leitura dos dados em um DeteFrame em Pandas
2. Análise básica dos conteúdos de cada coluna com contagem de valores
3. Elaboração de uma hipóteses que seja de interesse pessoal ou público com a utilização das variáveis do banco de dados, conforme o exemplo visto em aula (os homens ganham mais que as mulheres?) 
4. Tratamento dos dados com o objetivo de eliminar *outliers*, viés ou qualquer outro fator que atrapalhe a aleatoriedade de análise, para cada uma das variáveis, com visualização dos dados por heitograma ou outro tipo de gráfico
5. Definição das variáveis que serão utilizadas na análise
6. Cruzamento dos dados por máscara ou outra forma filtragam e separação, dentro da necessidade e interesse da proposta
7. Visualização dos dados cruzados para compreeensão das possíveis respostas (ver pyplot no final da segunda aula de Pandas = pandas102)
8. Direcionamento da problemática inicial através dos resultados das análises

### observação importante: para os dados da coluna estado civil, vamos considerar os valores 0.0 como solteiro e 1.0 como casado, para enriquecer as nossas possibilidades de análise.

## O Sistema SIDRA do IBGE

<br>
<img src="img/IBGE_SIDRA.png">
<br>

O Sistema IBGE de Recuperação Automática – SIDRA – visa facilitar aos administradores públicos e à sociedade em geral, através da Internet, a obtenção gratuita dos dados agregados de estudos e pesquisas realizados pelo IBGE.

Com o SIDRA é possível consultar dados na forma de séries temporais, acompanhando seu comportamento ao longo do tempo, bem como ter os mesmos disponibilizados por níveis territoriais desagregados, como município, distrito e bairro, de modo a facilitar o conhecimento da realidade municipal.

O sistema permite ao usuário:

- consulta universal através de navegador (browser) pela Internet;
- pesquisa por palavra-chave, temas etc;
- personalização das formas de apresentação das tabelas;
- visualização da tabela ou geração em arquivo;
- geração de gráficos a partir dos dados tabulares;
- envio de dados por e-mail a partir de agendamentos;
- visualização espacial dos dados através da geração de cartogramas;
- recuperação de séries temporais; 
- funcionamento integral durante todos os dias da semana;
- facilidade de reprodução por qualquer órgão da administração pública que queira adotar uma estratégia tabular para a disseminação de seus dados agregados pela Internet ou Intranet.

Atualmente estão disponíveis planos tabulares de 33 pesquisas, totalizando quase 600 milhões de dados agregados. O processo de disponibilização de novos dados é contínuo, o que faz com que cresça, ao longo do tempo, a cobertura das informações passíveis de serem extraídas a partir destes dados.

O SIDRA abrange:

- 33 pesquisas
- 371 variáveis agregados
- 972 tabelas
- 615 milhões de variáveis

E apresenta os seguintes temas:

- território
- população
- índices de preços
- emprego

- indústria
- comércio
- previsão de safra
- agricultura
- pecuária
- silvicultura
- horticultura
- extração vegetal
- orçamento familiar
- contas nacionais
- registro civil
- serviços
- empresas
